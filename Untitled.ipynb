````python
from lxml import etree
from collections import Counter, defaultdict
import itertools
from itertools import count
import networkx as nx
from frozendict import frozendict
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import xxhash

DBLP_FILE = 'data/dblp.xml'
DBLP_FILE_DTD = 'data/dblp.dtd'
````

````python
def iter_articles(year_range=None):
    """year_range is a range() object"""
    with open(DBLP_FILE, 'br') as fp, open(DBLP_FILE_DTD) as val:
        parser = etree.XMLParser(dtd_validation=True, no_network=True)
        context = etree.iterparse(fp, events=('end',), no_network=False, load_dtd=True)
        for action, elem in context:
            if elem.tag == 'article':
                if year_range:
                    year = elem.find('year')
                    if year is not None:
                        if int(year.text) in year_range:
                            yield elem
                else:
                    yield elem
            if elem.tag in ['inproceedings', 'proceedings', 'book', 'incollection',
                            'phdthesis', 'mastersthesis', 'www', 'person', 'data', 'article']:
                elem.clear()
````

````python
G = nx.Graph()
for elem in iter_articles(year_range=range(1980,1991)):
        article = frozendict({'type': 'article', 'title': elem.find('title').text, 'key': elem.get('key')})
        authors = [frozendict({'type': 'author', 'name': x.text}) for x in elem.findall('author')]
        G.add_node(article)
        G.add_edges_from([x, article] for x in authors)
        elem.clear()
AUTHORSHIP_GRAPH = G
````

````python
mapping = defaultdict(count().__next__)
G = nx.Graph()

with open('temp.txt', 'w') as fp:
    first = True
    for elem in iter_articles(year_range=range(1980,1991)):
            authors = [mapping[x.text] for x in elem.findall('author')]
            year = int(elem.find('year').text)

            for a,b in itertools.permutations(authors, 2):
                if not first:
                    fp.write('\n')
                first = False
                fp.write('{} {} {}'.format(a,b,year))
                     
TEMPORAL_COAUTHORSHIP_GRAPH = G
````

````python
count = Counter()
for elem in iter_articles():
    year = elem.find('year')
    if year is not None:
        year = int(year.text)
    else:
        continue
    count[year] += 1
````

````python
sns.set_style("whitegrid")
data = sorted(count.items(), key=lambda x: x[0])
df = pd.DataFrame(data={'year': [a for a,_ in data], 'count': [b for _,b in data]})
plt.subplots(figsize=(20,15))
g = sns.barplot(x="year", y="count", data = df)
plt.xticks(rotation=90)
plt.show()
````
